### Slicing
 - Maneira de fatiar uma lista ou array em vez de trabalhar com o valor completo

#### Slicing Unidimensionais
##### Series ou Arays
```python
  sequencia[início:fim:passo]

  ## Exemplo 1
  numeros = [10, 20, 30, 40, 50, 60, 70, 80, 90]
  
  # Para pegar os 3 primeiros itens da lista

  numeros[:3]

  OUTPUT -> [10,20,30]

  ## Exemplo 2

  # Para pegar a partir de um item da lista

  numeros[3:]

  # Inicio no item 3

  OUTPUT -> [40,50,60,70,...]

  ## Exemplo 3

  # pegar um item com intervalo

  numero [1:5:2]

  OUTPUT -> [20,40,60]

```

#### Slicing Multidimensional
##### Dataframes ou Arrays

### Interpolacao = Media dos valores anterior e posterior

````python
def interpolar(lista):
  if isinstance(lista, np.ndarray) == False:
    raise TypeError("O argumento deve ser um array do NumPy.")
  else:
    for i in range(len(lista)):
      if np.isnan(lista[i]):
        if i == 0:
          lista[i] = lista[i+1]
        elif i == len(lista)-1:
          lista[i] = lista[i-1]
        else:
          lista[i] = (lista[i-1] + lista[i+1])/2
    return lista
````

````python
Kaliningrad = interpolar(Kaliningrad)
````

````python
plt.plot(datas, Kaliningrad)
plt.plot(datas, Moscow)
plt.legend(['Kaliningrad','Moscow'])
print(f'Media Kaliningrad {np.mean(Kaliningrad)}')
print(f'Media Moscow {np.mean(Moscow)}')
````

## Comparando o Desempenho de um Array Numpy e uma lista do Python

````python
import time
````

````python
lista = list(range(1000000))
array = np.array(lista)
````

````python
start_time = time.time()

#elevar ao quadrado cada item da lista
lista_quadrado = [i**2 for i in lista]
lista_tempo = time.time() - start_time

print(f'Tempo da lista {lista_tempo}')
````

````python
start_time = time.time()

array_quadrado = array**2

array_tempo = time.time() - start_time

print(f'Tempo do Array {array_tempo}')
````

````python
if lista_tempo > array_tempo:
  diferenca_lista_array = lista_tempo / array_tempo
  porcentagem = (diferenca_lista_array) * 100
  print(f'O Array foi {porcentagem:.2f}% mais rápido')
else:
  diferenca_lista_array = array_tempo / lista_tempo
  porcentagem = (diferenca_lista_array) * 100
  print(f'A lista foi {porcentagem:.2f}% mais rápido')
````
